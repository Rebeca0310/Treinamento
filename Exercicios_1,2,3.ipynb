Crie um programa em Python que leia uma lista de nomes (Strings) informada pelo usuário. A entrada de nomes deve continuar até que o usuário digite a palavra “PARA”, que encerrará a leitura. Após isso, o programa deve solicitar uma letra para filtrar os nomes. Em seguida, exiba apenas os nomes que começam com essa letra, ignorando diferenças entre maiúsculas e minúsculas, e mostre o resultado em ordem alfabética.

Por exemplo, se a letra filtrada → A = Alice, Amanda, Ana.

```python
nomes_list = [] .

while True:
  nomes = str(input("Digite um nome (ou 'PARA' para encerrar): "))

  if nomes.upper() == "PARA":
    break
  nomes_list.append(nomes)

filtro_letras = str(input("Digite uma letra para filtrar os nomes: "))

nomes_filtro = []

for name in nomes_list:
  if name.lower().startswith(filtro_letras.lower()):
    nomes_filtro.append(name)

nomes_filtro.sort()

print("Nomes filtrados:")
if nomes_filtro:
  print(", ".join(nomes_filtro))
else:
  print("Nenhum nome encontrado com a letra informada.")
```

Crie um programa que funcione como um mini dicionário de traduções.

O usuário deve poder adicionar palavras em inglês e suas traduções em português, e depois consultar a tradução de uma palavra digitada.

Exemplo:

Digite a palavra em inglês (ou 'sair'): car

Digite a tradução em português: carro

Digite a palavra em inglês (ou 'sair'): house

Digite a tradução em português: casa

Digite a palavra em inglês para buscar: house

Tradução: casa






```python
dicionario_traducoes = {}
while True:
  usuario = str(input("Digite a palavra em inglês (ou 'sair'): "))
  if usuario.lower() == "sair":
    break
  traducao = str(input("Digite a tradução em português: "))
  dicionario_traducoes[usuario] = traducao

palavra_usuario = str(input("Digite a palavra em inglês para buscar: "))
print(f"Tradução: {dicionario_traducoes.get(palavra_usuario, 'Nenhuma tradução encontrada para essa palavra.')}")
```

Tuplas são muito parecidas com listas! A grande diferença é que, uma vez criada a tupla, ela não pode mais mudar!! Ou seja, nada de acrescentar novos elemento, remover ou alterar elementos.


    Ápos a Tupla ser criada, os dados não podem ser modificados

    Tuplas são mais rapidas na execução por serem multaveis.

    Como as Tuplas não podem ser modificadas, elas são consideradas mais seguras do que as listas.
