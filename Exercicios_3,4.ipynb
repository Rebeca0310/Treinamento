Você está desenvolvendo um sistema de controle de acesso. Considere as permissões obrigatórias para acessar uma funcionalidade:



    required_permissions = {"read", "write", "execute", "admin"}
e as permissões atualmente atribuídas ao usuário:



    user_permissions = {"read", "write", "delete", "delete"}
Verifique se o usuário possui todas as permissões necessárias. Caso não possua, identifique quais permissões estão faltando e se há duplicatas na lista de permissões do usuário.

```python
required_permissions = {"read", "write", "execute", "admin"}
user_permissions = {"read", "write", "delete", "delete"}

tem_duplicatas = len(user_permissions) != len(set(user_permissions))

if tem_duplicatas:
    print("Há duplicatas na lista de permissões do usuário.")
else:
    faltando = required_permissions - user_permissions
    if faltando:
        print("O usuário está faltando as seguintes permissões:", faltando)
    else:
        print("O usuário possui todas as permissões necessárias.")
```

Crie um sistema de terminal para gerenciar personagens de RPG.

Cada personagem deve possuir os seguintes campos:

* Nome

* Classe

* Nível

* Inventário

O inventário deve armazenar vários itens.

**Menu**

* Adicionar personagem

* Consultar personagem

* Adicionar item ao inventário

* Remover item do inventário

* Listar personagens

* Sair

**Regras**
Adicionar personagem
Solicite:

* Nome

* Classe

* Nível

**Restrições:**

    Não permitir personagens com nomes duplicados.

    O inventário deve iniciar vazio.

    O nível não pode ser negativo.

**Consultar personagem**

* Solicite o nome do personagem.

* Exiba todas as informações cadastradas do personagem.

* Caso ele não exista, informe ao usuário.

**Adicionar item ao inventário**

Solicite:

* Nome do personagem

* Nome do item

Adicione o item ao inventário do personagem.

Caso o personagem não exista, informe ao usuário.

**Remover item do inventário**
Solicite:

* Nome do personagem

* Nome do item

Remova o item apenas se ele existir no inventário.

Caso o personagem ou item não existam, informe ao usuário.

**Listar personagens**

Exiba o nome, classe e nível de todos os personagens cadastrados.

Exemplo:

      === PERSONAGENS ===

      Arthur
      Classe: Cavaleiro
      Nível: 10

      Merlin
      Classe: Mago
      Nível: 20

```python
personagens = {}

def adicionar_personagem(personagens):
    nome = input("Digite o nome do personagem: ").strip()
    if nome in personagens:
        print("Personagem já existe.")
        return

    classe = input("Digite a classe do personagem: ").strip()
    try:
        nivel = int(input("Digite o nível do personagem: "))
        if nivel < 0:
            print("Nível não pode ser negativo.")
            return
    except ValueError:
        print("Nível inválido. Deve ser um número.")
        return

    personagens[nome] = {
        "classe": classe,
        "nivel": nivel,
        "inventario": []
    }
    print(f"Personagem {nome} adicionado com sucesso.")


def consultar_personagem(personagens):
    nome = input("Digite o nome do personagem: ").strip()
    if nome not in personagens:
        print("Personagem não encontrado.")
        return

    personagem = personagens[nome]
    print(f"\nNome: {nome}")
    print(f"Classe: {personagem['classe']}")
    print(f"Nível: {personagem['nivel']}")
    print(f"Inventário: {', '.join(personagem['inventario']) if personagem['inventario'] else 'Vazio'}")


def adicionar_item_inventario(personagens):
    nome = input("Nome do personagem: ").strip()

    if nome in personagens:
        item = input("Nome do item: ").strip()
        personagens[nome]["inventario"].append(item)
        print(f"{item} adicionado ao inventário de {nome}!")
    else:
        print(f"Personagem {nome} não encontrado.")


def remover_item_inventario(personagens):
    nome = input("Nome do personagem: ").strip()

    if nome not in personagens:
        print(f"Personagem {nome} não encontrado.")
        return

    item = input("Nome do item: ").strip()
    if item in personagens[nome]["inventario"]:
        personagens[nome]["inventario"].remove(item)
        print(f" {item} removido do inventário de {nome}!")
    else:
        print(f"O item '{item}' não está no inventário de {nome}.")


def listar_personagens(personagens):
    print("\n=== PERSONAGENS ===")
    for nome, personagem in personagens.items():
        print(f"\n{nome}")
        print(f"Classe: {personagem['classe']}")
        print(f"Nível: {personagem['nivel']}")

def menu_principal():
    while True:
        print("\n1. Adicionar personagem")
        print("2. Consultar personagem")
        print("3. Adicionar item ao inventário")
        print("4. Remover item do inventário")
        print("5. Listar personagens")
        print("6. Sair")

        opcao = input("Escolha uma opção: ").strip()

        if opcao == "1":
            adicionar_personagem(personagens)
        elif opcao == "2":
            consultar_personagem(personagens)
        elif opcao == "3":
            adicionar_item_inventario(personagens)
        elif opcao == "4":
            remover_item_inventario(personagens)
        elif opcao == "5":
            listar_personagens(personagens)
        elif opcao == "6":
            print("Saindo do programa.")
            break
        else:
            print("Opção inválida. Tente novamente.")

# Execução do programa
if __name__ == "__main__":
    menu_principal()
```
